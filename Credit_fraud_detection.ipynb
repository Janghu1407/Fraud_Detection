```python
def print_results(true, pred):
    print('accuracy_score: {}'.format(accuracy_score(true, pred)))
    print('precision_score: {}'.format(precision_score(true, pred)))
    print('recall_score: {}'.format(recall_score(true, pred)))
    print('f1_score: {}'.format(f1_score(true, pred)))
```

```python
x = data.drop('Class', axis=1)
y = data['Class']
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, random_state = 7, test_size=0.2)
```

```python
log = LogisticRegression(C=0.8)
```

```python
log.fit(x_train, y_train)
```

```python
log_pred = log.predict(x_test)
```

```python
confusion_matrix(y_test, log_pred)
```

```python
print_results(y_test, log_pred)
```

```python
plot_precision_recall_curve(log, x_test, y_test)
```

```python
svc = SVC(C=10, gamma=0.1, kernel='rbf')
```

```python
svc.fit(x_train,y_train)
```

```python
svc_pred = svc.predict(x_test)
```

```python
confusion_matrix(y_test, svc_pred)
```

```python
print_results(y_test, svc_pred)
```

```python
svc1 = SVC()
```

```python
svc1.fit(x_train,y_train)
```

```python
svc1_pred = svc1.predict(x_test)
```

```python
confusion_matrix(y_test, svc1_pred)
```

```python
print_results(y_test, svc1_pred)
```

```python
plot_precision_recall_curve(svc1, x_test, y_test)
```

```python
rf = RandomForestClassifier(max_depth= None, max_features=8, n_estimators=20)
```

```python
rf.fit(x_train, y_train)
```

```python
rf_pred = rf.predict(x_test)
```

```python
confusion_matrix(y_test, rf_pred)
```

```python
print_results(y_test, rf_pred)
```

```python
plot_precision_recall_curve(rf, x_test, y_test)
```
